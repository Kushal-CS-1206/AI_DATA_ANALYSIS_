```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'X': [5, 15, 25, 35, 45, 55],
    'Y': [3, 14, 24, 36, 47, 53]
}

df = pd.DataFrame(data)

# Scatter plot
sns.scatterplot(x='X', y='Y', data=df, color='teal')
plt.title('Relationship Between X and Y')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.grid(True)
plt.show()
#
```
