```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
data = np.random.rand(6, 6)
columns = ['A', 'B', 'C', 'D', 'E', 'F']
df = pd.DataFrame(data, columns=columns)

# Correlation matrix
corr = df.corr()

# Heatmap with advanced styling
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, linewidths=1, linecolor='white', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix Heatmap')
plt.show()
#
```
