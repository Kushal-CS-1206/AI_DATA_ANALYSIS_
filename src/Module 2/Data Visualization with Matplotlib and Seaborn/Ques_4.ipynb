```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample sales data
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [150, 200, 180, 220, 170]
}

df = pd.DataFrame(data)

# Bar plot
sns.barplot(x='Product', y='Sales', data=df, palette='pastel')
plt.title('Sales Comparison by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()







```
