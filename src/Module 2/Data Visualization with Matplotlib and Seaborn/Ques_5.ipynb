```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weigh
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76]
}

df = pd.DataFrame(data)

# Scatter plot
sns.scatterplot(x='Height', y='Weight', data=df, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```
