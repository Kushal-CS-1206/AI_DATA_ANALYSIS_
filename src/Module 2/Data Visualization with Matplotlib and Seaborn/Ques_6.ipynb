```python

import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [150, 200, 180, 220, 170]

# Plot customization
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue')
plt.title('Sales Comparison by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product', fontsize=12, fontstyle='italic')
plt.ylabel('Sales', fontsize=12, fontstyle='italic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
