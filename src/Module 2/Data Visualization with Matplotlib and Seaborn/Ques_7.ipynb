```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
}

df = pd.DataFrame(data)

# Enhanced scatter plot
sns.scatterplot(x='Height', y='Weight', size='Size', data=df, legend=False, sizes=(50, 300), color='purple')
plt.title('Height vs Weight with Size Variation')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
##
```
