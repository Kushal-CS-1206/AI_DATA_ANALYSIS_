```python

import seaborn as sns
import matplotlib.pyplot as plt

# Data
marks = [23, 45, 56, 78, 89, 90, 67, 88, 76, 59]

# Histogram with KDE
sns.histplot(marks, kde=True, color='orange', bins=5)
plt.title('Distribution of Marks with KDE')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
#
```
