```python
#
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [150, 200, 180, 220, 170]
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Sales Comparison by Product')
axs[0].set_xlabel('Product')
axs[0].set_ylabel('Sales')
axs[0].grid(True, axis='y', linestyle='--', alpha=0.7)

# Scatter plot
axs[1].scatter(heights, weights, color='green')
axs[1].set_title('Height vs Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
