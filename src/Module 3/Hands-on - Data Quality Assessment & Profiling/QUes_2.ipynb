```python
import pandas as pd
import seaborn as sns

# Part 1: Load Titanic Dataset and Check Missing Values.
titanic = sns.load_dataset('titanic')

# Check missing values in 'age' column
missing_age_count = titanic['age'].isnull().sum()
print(f"Missing values in 'age' column: {missing_age_count}")

# Part 2: Identify Duplicate Rows in Titanic Dataset
duplicate_rows = titanic[titanic.duplicated()]
print(f"\nNumber of duplicate rows: {len(duplicate_rows)}")
print(duplicate_rows)

# Part 3: Generate Data Quality Report
missing_values = titanic.isnull().sum()
num_duplicates = titanic.duplicated().sum()
basic_stats = titanic.describe(include='all')

print("\nMissing values per column:")
print(missing_values)

print(f"\nNumber of duplicate rows: {num_duplicates}")

print("\nBasic statistics:")
print(basic_stats)
```
