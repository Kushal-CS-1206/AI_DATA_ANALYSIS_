```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt

# 1. Missing Data Examples

# Task 1: Customer emails missing count
csv_customers = """
CustomerID,Name,Email
1,John,john@example.com
2,Alice,
3,Bob,bob@example.com
4,Diana,diana@example.com
5,Evan,
"""
df_customers = pd.read_csv(StringIO(csv_customers))
incomplete_emails = df_customers['Email'].isnull().sum()
print(f"Missing customer emails: {incomplete_emails}")

# Task 2: Sales dataset with missing transaction dates percentage
csv_sales = """
TransactionID,Date,Amount
1001,2023-01-01,200
1002,,150
1003,2023-01-03,300
1004,,400
1005,2023-01-05,250
"""
df_sales = pd.read_csv(StringIO(csv_sales))
missing_dates_pct = df_sales['Date'].isnull().mean() * 100
print(f"Percentage of missing transaction dates: {missing_dates_pct:.2f}%")

# Task 3: Missing department in employee registry
csv_employees = """
EmployeeID,Name,Department
1,John,Sales
2,Alice,Marketing
3,Bob,
4,Diana,HR
5,Evan,
"""
df_employees = pd.read_csv(StringIO(csv_employees))
missing_dept = df_employees['Department'].isnull().sum()
print(f"Missing department entries: {missing_dept}")

# 2. Duplicate Data Examples

# Task 1: Duplicate entries in customer dataset
csv_dup_customers = """
CustomerID,Name,Email
1,John,john@example.com
2,Alice,alice@example.com
3,Bob,bob@example.com
2,Alice,alice@example.com
4,Diana,diana@example.com
"""
df_dup_customers = pd.read_csv(StringIO(csv_dup_customers))
num_duplicates = df_dup_customers.duplicated().sum()
print(f"Number of duplicate customer records: {num_duplicates}")

# Task 2: Varying date formats example
csv_dates = """
ID,Date
1,12/31/2023
2,31/12/2023
3,01/15/2024
4,15/01/2024
"""
df_dates = pd.read_csv(StringIO(csv_dates))
print("\nDate column samples with different formats:")
print(df_dates['Date'])

# Task 3: Varying phone number formats
csv_phones = """
ID,Phone
1,(123) 456-7890
2,123-456-7890
3,1234567890
4,123.456.7890
"""
df_phones = pd.read_csv(StringIO(csv_phones))
print("\nPhone number formats:")
print(df_phones['Phone'])

# Task 4: State abbreviations inconsistencies
csv_addresses = """
ID,State
1,CA
2,Calif.
3,NY
4,New York
"""
df_addresses = pd.read_csv(StringIO(csv_addresses))
print("\nState entries with inconsistencies:")
print(df_addresses['State'])

# 4. Data Drift Examples

# Task 1: Monthly revenues over six months
csv_revenue = """
Month,Revenue
Jan,10000
Feb,11000
Mar,10500
Apr,12000
May,15000
Jun,20000
"""
df_revenue = pd.read_csv(StringIO(csv_revenue))
plt.plot(df_revenue['Month'], df_revenue['Revenue'], marker='o')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# Task 2: User engagement over quarters
csv_engagement = """
Quarter,ActiveUsers
Q1,5000
Q2,5200
Q3,4800
Q4,4700
"""
df_engagement = pd.read_csv(StringIO(csv_engagement))
plt.bar(df_engagement['Quarter'], df_engagement['ActiveUsers'], color='skyblue')
plt.title("User Engagement Over Quarters")
plt.xlabel("Quarter")
plt.ylabel("Active Users")
plt.show()

# Task 3: Stock price anomalies over a year
csv_stock = """
Month,Price
Jan,150
Feb,155
Mar,160
Apr,158
May,170
Jun,300
Jul,165
Aug,160
Sep,155
Oct,150
Nov,145
Dec,140
"""
df_stock = pd.read_csv(StringIO(csv_stock))
plt.plot(df_stock['Month'], df_stock['Price'], marker='x', color='red')
plt.title("Stock Price Over Year (Anomaly in June)")
plt.xlabel("Month")
plt.ylabel("Price")
plt.show()



```
