```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
import re

# 1. Descriptive.

# Task 1: Sales figures stats
csv_sales = """
SaleID,Sales
1,200
2,220
3,210
4,230
5,240
"""
df_sales = pd.read_csv(StringIO(csv_sales))
mean_sales = df_sales['Sales'].mean()
median_sales = df_sales['Sales'].median()
mode_sales = df_sales['Sales'].mode()[0]
print(f"Sales - Mean: {mean_sales}, Median: {median_sales}, Mode: {mode_sales}")

# Task 2: Customer demographics (age)
csv_customers = """
CustomerID,Age
1,25
2,30
3,25
4,40
5,30
"""
df_customers = pd.read_csv(StringIO(csv_customers))
mean_age = df_customers['Age'].mean()
median_age = df_customers['Age'].median()
mode_age = df_customers['Age'].mode()[0]
print(f"Age - Mean: {mean_age}, Median: {median_age}, Mode: {mode_age}")

# Task 3: Website visit counts
csv_visits = """
Day,Visits
Mon,100
Tue,120
Wed,100
Thu,130
Fri,120
"""
df_visits = pd.read_csv(StringIO(csv_visits))
mean_visits = df_visits['Visits'].mean()
median_visits = df_visits['Visits'].median()
mode_visits = df_visits['Visits'].mode()[0]
print(f"Visits - Mean: {mean_visits}, Median: {median_visits}, Mode: {mode_visits}")


# 2. Distribution Analysis

# Task 1: Product prices histogram
csv_prices = """
ProductID,Price
1,10.5
2,15.0
3,12.5
4,20.0
5,18.5
"""
df_prices = pd.read_csv(StringIO(csv_prices))
plt.hist(df_prices['Price'], bins=5, color='skyblue')
plt.title("Product Prices Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Task 2: Exam scores distribution
csv_scores = """
StudentID,Score
1,78
2,85
3,90
4,88
5,76
"""
df_scores = pd.read_csv(StringIO(csv_scores))
sns.histplot(df_scores['Score'], kde=True, color='orange')
plt.title("Exam Scores Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()

# Task 3: Order quantities distribution
csv_orders = """
OrderID,Quantity
1,5
2,10
3,7
4,8
5,12
"""
df_orders = pd.read_csv(StringIO(csv_orders))
plt.hist(df_orders['Quantity'], bins=5, color='green')
plt.title("Order Quantities Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()


# 3. Data Type Validation

# Task 1: Validate numeric fields
csv_numeric = """
ID,Value
1,100
2,200
3,abc
4,400
5,500
"""
df_numeric = pd.read_csv(StringIO(csv_numeric))
numeric_validation = pd.to_numeric(df_numeric['Value'], errors='coerce')
invalid_numeric = numeric_validation.isnull().sum()
print(f"Invalid numeric entries: {invalid_numeric}")

# Task 2: Check valid date formats
csv_dates = """
ID,Date
1,2023-01-01
2,2023/01/02
3,01-03-2023
4,2023-13-01
5,2023-01-05
"""
df_dates = pd.read_csv(StringIO(csv_dates))
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str, format='%Y-%m-%d', errors='raise')
        return True
    except:
        return False
valid_dates = df_dates['Date'].apply(is_valid_date)
invalid_dates_count = (~valid_dates).sum()
print(f"Invalid date formats: {invalid_dates_count}")

# Task 3: Validate email formats
csv_emails = """
ID,Email
1,john@example.com
2,alice@@example.com
3,bob.example.com
4,diana@example.com
5,evan@example
"""
df_emails = pd.read_csv(StringIO(csv_emails))

email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
valid_email = df_emails['Email'].str.match(email_pattern)
invalid_email_count = (~valid_email).sum()
print(f"Invalid email formats: {invalid_email_count}")







```
