```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Sample data with numeric attribute 'sales' for two time periods
csv_data = """
period,sales
2023-01,200
2023-01,220
2023-01,210
2023-02,230
2023-02,240
2023-02,250
2024-01,300
2024-01,320
2024-01,310
2024-02,330
2024-02,340
2024-02,350
"""

df = pd.read_csv(StringIO(csv_data))

# Split data into two periods (2023 and 2024)
df['year'] = df['period'].str[:4]
sales_2023 = df[df['year'] == '2023']['sales']
sales_2024 = df[df['year'] == '2024']['sales']

# Descriptive statistics
print("2023 Sales stats:")
print(sales_2023.describe())
print("\n2024 Sales stats:")
print(sales_2024.describe())

# Plot histograms to compare distributions
plt.hist(sales_2023, bins=5, alpha=0.5, label='2023 Sales')
plt.hist(sales_2024, bins=5, alpha=0.5, label='2024 Sales')
plt.legend()
plt.title("Sales Distribution Comparison")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()








```
