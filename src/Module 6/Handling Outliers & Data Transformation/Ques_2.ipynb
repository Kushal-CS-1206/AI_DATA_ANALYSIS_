```python
import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt

# Sample dataset with outliers
np.random.seed(0)
data = np.concatenate([np.random.normal(50, 10, 100), np.array([10, 100])])
df = pd.DataFrame({'value': data})

# Apply Winsorization (limit 5% on both tails)
winsorized_data = winsorize(df['value'], limits=[0.05, 0.05])
df['winsorized'] = winsorized_data

# Plot original vs winsorized data
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.boxplot(df['value'])
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.boxplot(df['winsorized'])
plt.title("Winsorized Data")

plt.tight_layout()
plt.show()
```
