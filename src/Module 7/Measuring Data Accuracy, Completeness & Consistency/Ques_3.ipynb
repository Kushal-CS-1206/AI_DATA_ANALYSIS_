## Compare Data Completeness Over Time

**Description**: Analyze the trend of missing data in `"sales_data.csv"` over several months stored in a "date" column. Visualize missing data rates by month.

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Sample dataset with missing values and dates
sales_csv = """date,product_id,sales,region
2023-01-05,101,200,North
2023-01-15,102,,South
2023-02-03,103,150,West
2023-02-18,104,,
2023-03-01,105,250,East
2023-03-22,106,,North
2023-04-10,107,300,
2023-04-15,108,275,West
"""

# Load the dataset
df = pd.read_csv(StringIO(sales_csv), parse_dates=['date'])

# Add a month column
df['month'] = df['date'].dt.to_period('M')

# Calculate missing data rate per month
missing_rates = df.isnull().groupby(df['month']).mean() * 100

# Plot
missing_rates.plot(kind='bar', figsize=(10, 6))
plt.title("Monthly Missing Data Rate (%)")
plt.ylabel("Missing Rate (%)")
plt.xlabel("Month")
plt.legend(title="Column")
plt.tight_layout()
plt.show()
```
