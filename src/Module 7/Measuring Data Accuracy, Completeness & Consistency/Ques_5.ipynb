## Validate Data Completeness for Critical Fields

**Description**: In `transaction_data.csv` , calculate the missing percentage of critical fields: "transaction_id", "amount", and "date", and visualize them.

```python
# Write your code from here
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Sample transaction_data.csv content
transaction_data_csv = """transaction_id,amount,date,customer_id
1001,250.75,2023-01-15,501
1002,,2023-01-16,502
,300.00,2023-01-17,503
1004,150.00,,504
1005,200.00,2023-01-19,505
"""

# Load dataset
transaction_data = pd.read_csv(StringIO(transaction_data_csv))

# Define critical fields
critical_fields = ["transaction_id", "amount", "date"]

# Calculate missing percentage for each critical field
missing_percent = transaction_data[critical_fields].isna().mean() * 100

# Print missing percentage
print("Missing percentage for critical fields:")
print(missing_percent)

# Visualization
missing_percent.plot(kind='bar', color='tomato')
plt.title("Missing Data Percentage in Critical Fields")
plt.ylabel("Missing Percentage (%)")
plt.xlabel("Fields")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
