### Detecting & Handling Imbalanced Data: Visualizing Class Imbalance
**Question**: Load the Credit Card Fraud Detection dataset and visualize the class imbalance. Then apply random undersampling to balance it.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Sample dataset creation (since the real dataset is large)
data = {
    'feature1': range(1000),
    'feature2': range(1000, 2000),
    'class': [0]*980 + [1]*20  # Imbalanced: 980 normal, 20 fraud
}
df = pd.DataFrame(data)

# Visualize original class distribution
df['class'].value_counts().plot(kind='bar', title='Original Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Separate majority and minority classes
df_majority = df[df['class'] == 0]
df_minority = df[df['class'] == 1]

# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                   replace=False,
                                   n_samples=len(df_minority),
                                   random_state=42)

# Combine minority class with downsampled majority class
df_balanced = pd.concat([df_majority_downsampled, df_minority])

# Visualize balanced class distribution
df_balanced['class'].value_counts().plot(kind='bar', title='Balanced Class Distribution (Random Undersampling)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
#
```
