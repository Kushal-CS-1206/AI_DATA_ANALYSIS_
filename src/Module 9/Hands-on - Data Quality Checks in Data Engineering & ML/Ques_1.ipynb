### Detect Data Drift using Kolmogorov-Smirnov Test
**Description**: Use statistical tests to detect data drift between two datasets.

```python
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

# Simulate training data (n
np.random.seed(42)
train_data = np.random.normal(loc=50, scale=10, size=1000)

# Simulate production data with drift (shifted distribution)
production_data = np.random.normal(loc=60, scale=10, size=1000)

# Apply Kolmogorov-Smirnov test
ks_stat, p_value = ks_2samp(train_data, production_data)

# Print results
print("Kolmogorov-Smirnov Test:")
print(f"KS Statistic: {ks_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("Data drift detected (statistically significant difference).")
else:
    print("No significant data drift detected.")

# Plot for visual inspection
plt.hist(train_data, bins=30, alpha=0.5, label='Train Data')
plt.hist(production_data, bins=30, alpha=0.5, label='Production Data')
plt.legend()
plt.title('Distribution Comparison: Train vs. Production')
plt.show()
```
