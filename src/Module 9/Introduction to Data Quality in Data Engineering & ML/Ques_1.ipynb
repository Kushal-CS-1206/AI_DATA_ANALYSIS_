### Detect Data Drift
**Description**: Data drift can occur when the statistical properties of your data change over time. Learn to detect data drift using visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate training data distribution
np.random.seed(42)
train_data = pd.DataFrame({
    "feature1": np.random.normal(loc=50, scale=5, size=1000),
    "feature2": np.random.normal(loc=30, scale=3, size=1000)
})

# Simulate new data with drift in feature1 (mean shifted)
new_data = pd.DataFrame({
    "feature1": np.random.normal(loc=60, scale=5, size=1000),  # mean shifted
    "feature2": np.random.normal(loc=30, scale=3, size=1000)
})

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.kdeplot(train_data["feature1"], label="Train", ax=axs[0])
sns.kdeplot(new_data["feature1"], label="New", ax=axs[0])
axs[0].set_title("Feature1 Distribution Drift")
axs[0].legend()

sns.kdeplot(train_data["feature2"], label="Train", ax=axs[1])
sns.kdeplot(new_data["feature2"], label="New", ax=axs[1])
axs[1].set_title("Feature2 Distribution (No Drift)")
axs[1].legend()

plt.show()
```
